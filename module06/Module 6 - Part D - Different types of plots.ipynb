**Import the matplotlib library:**

```python
import matplotlib.pyplot as plt # Import the matplotlib library
import numpy as np # We will use numpy to create some data

from IPython.display import display # Import display function from IPython (so we can edit the figures as we go)
```

**Line graph**

```python
fig = plt.figure() # Create a figure object

ax = fig.add_subplot() # Add axes to the figure

X1 = [1, 2, 3] # Define X data
Y1 = [4, 5, 6] # Define Y data

ax.set_xlabel('X Axis Label') # Set the name of the x-axis
ax.set_ylabel('Y Axis Label') # set the name of the y-axis
ax.set_title('Line graph example')  # Set the title of the plot

ax.plot(X1, Y1) # plotting a line graph
```

**Scatter plot**

```python

fig = plt.figure()
ax = fig.add_subplot()
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_title('Scatter Plot Example')

# Generate scatter plot data with an upward right trend
x_trend = np.linspace(0, 10, 50) # creating 50 equally spaced points from 0 to 10
y_trend = 2*x_trend + np.random.normal(0, 2, 50)  # creating y data with some noise

ax.scatter(x_trend, y_trend, color='blue', marker='o') # Command for creating a scatter plot


# Fit a linear trendline to the scatter plot data
coeffs = np.polyfit(x_trend, y_trend, deg=1) # deg=1 for linear fit
y_fit = np.polyval(coeffs, x_trend)

ax.plot(x_trend, y_fit, color='red', linestyle='--', label='Trendline')

plt.show()
```

**Bar chart**

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.set_xlabel('Category')
ax.set_ylabel('Value')
ax.set_title('Bar Chart Example')

# Generate sample data for the bar chart
categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(10, 100, size=len(categories)) # This will generate a random integer between 10 and 100 for each category 

ax.bar(categories, values, color='blue') # command for creating a bar chart
plt.show()
```

**Grouped bar chart**

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.set_xlabel('Category')
ax.set_ylabel('Value')
ax.set_title('Grouped Bar Chart Example')

# Main categories
categories = ['A', 'B', 'C', 'D', 'E']
x = np.arange(len(categories))

# Generate random values for 3 subcategories
values_sub1 = np.random.randint(10, 100, size=len(categories))
values_sub2 = np.random.randint(10, 100, size=len(categories))
values_sub3 = np.random.randint(10, 100, size=len(categories))

# Bar width and positions
bar_width = 0.25
ax.bar(x - bar_width, values_sub1, width=bar_width, color='blue', label='Subcategory 1')
ax.bar(x, values_sub2, width=bar_width, color='skyblue', label='Subcategory 2')
ax.bar(x + bar_width, values_sub3, width=bar_width, color='lightblue', label='Subcategory 3')

# Set x-axis tick positions and labels
ax.set_xticks(x)
ax.set_xticklabels(categories)

ax.legend()
plt.show()
```

**Combination plot**

```python
# Sample data
categories = ['A', 'B', 'C', 'D', 'E', 'F']
bar_values = np.random.randint(50, 100, size=len(categories))
line_values = np.random.uniform(0.3, 0.8, size=len(categories))

# Create figure and primary axis
fig, ax1 = plt.subplots(figsize=(8, 5))

# Plot bar chart on primary y-axis
bar_positions = np.arange(len(categories))
bar_width = 0.6
bars = ax1.bar(bar_positions, bar_values, width=bar_width, color='blue', label='Bars')

ax1.set_xlabel('Category')
ax1.set_ylabel('Bar values')
ax1.set_xticks(bar_positions)
ax1.set_xticklabels(categories)

# Create secondary y-axis
ax2 = ax1.twinx()
ax2.plot(bar_positions, line_values, color='red', marker='o', linestyle='-', linewidth=2, label='Line')
ax2.set_ylabel('Line values')

# Add grid, title, and combined legend ---
ax1.grid(axis='y', linestyle='--', alpha=0.6)
fig.suptitle("Combined Bar and Line Plot ")

plt.tight_layout()
plt.show()
```

**Normalized vs Raw values**

```python
fig, ax = plt.subplots(figsize=(7, 4))

base_raw = [9052, 6803, 8447, 4592, 7501, 6345]
A_raw = [8403, 5714, 7785, 4123, 7108, 6237]
B_raw = [7607, 5105, 7105, 3840, 6857, 5835]

X = [1, 2, 3, 4, 5, 6] # X data
ax.set_xlabel('Condition') # X axis label
ax.set_ylabel('Raw values') # Y axis label
ax.plot(X, base_raw, label='Baseline', marker='o')
ax.plot(X, A_raw, label='Alternative A', marker='o')
ax.plot(X, B_raw, label='Alternative B', marker='o')
plt.legend()
plt.show()

base_norm = [base_raw[i]/base_raw[i] for i in range(len(base_raw))]
A_norm = [A_raw[i]/base_raw[i] for i in range(len(base_raw))]
B_norm = [B_raw[i]/base_raw[i] for i in range(len(base_raw))]

fig, ax = plt.subplots(figsize=(7, 4))

ax.set_xlabel('Condition') # X axis label
ax.set_ylabel('Normalized values') # Y axis label
ax.plot(X, base_norm, label='Baseline', marker='o')
ax.plot(X, A_norm, label='Alternative A', marker='o')
ax.plot(X, B_norm, label='Alternative B', marker='o')
plt.legend()
plt.show()


```

**Pie chart**

```python
fig = plt.figure(figsize=(9, 7))
ax = fig.add_subplot()

# Generate sample data for the pie chart
pie_labels = ['A', 'B', 'C', 'D']
pie_sizes = np.random.randint(10, 50, size=len(pie_labels))
pie_colors = ['red', 'blue', 'green', 'orange']

wedges, texts, autotexts = ax.pie(pie_sizes, labels=pie_labels, colors = pie_colors, autopct='%1.1f%%', startangle=90) 

for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontsize(12)  # optional: change font size
    autotext.set_fontweight('bold')  # optional: make it bold

plt.show()
```

**Spider/Radar plot**

```python
fig = plt.figure()
ax_spider = fig.add_subplot(polar=True) #polar=True makes the plot circular
ax_spider.set_title('Spider Plot Example')

# Generate sample data for the spider plot
labels = ['A', 'B', 'C', 'D', 'E']
num_vars = len(labels)

values_spider = np.random.randint(10, 100, size=num_vars) # This will generate a random integer between 10 and 100 for each category 
values_spider = np.append(values_spider, values_spider[0]) # Repeat the first value to close the circle

# Calculate angle for each axis
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist() # this will create angles for each axis
angles += angles[:1] # this will close the circle

# Draw one axe per variable + add labels
ax_spider.set_xticks(angles[:-1])
ax_spider.set_xticklabels(labels)

# Plot data
ax_spider.plot(angles, values_spider, color='blue', linewidth=2) # Command for creating a spider plot
ax_spider.fill(angles, values_spider, color='blue', alpha=0.25) # Fill the area

plt.show()
```

**3D Surface Plot**

```python
# Generate grid data for 3D surface
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.cos(Y)  # Example surface function

fig = plt.figure(figsize=(9, 7))
ax = fig.add_subplot(projection='3d')
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')

surf = ax.plot_surface(X, Y, Z, cmap='inferno', edgecolor='none')

# Change the view
ax.view_init(elev=25, azim=125)

# Customize grid appearance
ax.xaxis._axinfo['grid']['color'] = 'lightgray'
ax.xaxis._axinfo['grid']['linestyle'] = '--'

ax.yaxis._axinfo['grid']['color'] = 'lightgray'
ax.yaxis._axinfo['grid']['linestyle'] = '--'

ax.zaxis._axinfo['grid']['color'] = 'lightgray'
ax.zaxis._axinfo['grid']['linestyle'] = '--'

# Add a color bar
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=10, pad=0.09)

plt.show()
```

**Wind/Wave rose**

A graphical representation often used within wind energy is the wind/wave rose. These are radar plots that show the frequency and magnitude of different wind or wave directions. We can install and import the 'windrose' library to help us with plotting a wind or wave rose

```python
from windrose import WindroseAxes # use pip install windrose to install the package

# Simulate wind mostly from the East (90°)
wind_directions = np.random.normal(loc=90, scale=100, size=500)  # mean=90°, std dev=100°
wind_directions = np.mod(wind_directions, 360)  # wrap around 0–360°

# Simulate random wind speeds
wind_speeds = np.random.uniform(0, 25, 500)

# Create wind rose plot
fig = plt.figure(figsize=(6, 6))
ax = WindroseAxes.from_ax(fig=fig)
ax.bar(wind_directions, wind_speeds, normed=True, opening=1, edgecolor='white')
ax.set_legend(title="Wind speed (m/s)")
plt.show()
```
