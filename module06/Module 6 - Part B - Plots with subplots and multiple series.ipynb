**Import the matplotlib library:**

```python
import matplotlib.pyplot as plt
import pandas as pd

plt.ioff()
```

**Let's recall from Lecture 1:**

```python
df = pd.read_excel("example_data.xlsx") # Read data from an Excel file into a DataFrame

X = df["wind speed"] # Extract the "wind speed" column for X data
Y = df["power"]      # Extract the "power" column for Y data

# Simple line plot
plt.plot(X, Y)
plt.xlabel("Wind speed (m/s)")
plt.ylabel("Power (kW)")
plt.title("Power Curve")
plt.show()
```

**Now, let's create a new plot with multiple data series**

```python
df = pd.read_excel("example_data.xlsx", sheet_name='Lecture 2')

X = df["wind speed"]
Y1 = df["power 1"] 
Y2 = df["power 2"]

fig, ax = plt.subplots()

ax.set_xlabel('Wind speed (m/s)')
ax.set_ylabel('Power (kW)')
ax.set_title('Power Curve')

ax.plot(X, Y1)
ax.plot(X, Y2)
plt.show()
```

**We can also add labels to our data series and add them to a legend:**

```python
# Adding labels to the lines

fig, ax = plt.subplots()

ax.set_xlabel('Wind speed (m/s)')
ax.set_ylabel('Power (kW)')
ax.set_title('Power Curve')

ax.plot(X, Y1, label='Turbine 1') 
ax.plot(X, Y2, label='Turbine 2')

ax.legend()  # Add a legend to distinguish the lines
plt.show()
```

**Multiple axes**

The y-axis magnitudes of our data series are quite different. We can use multiple y-axes to show them better.

```python
fig, ax1 = plt.subplots() # Create a figure and the first axes object

# Plot the first data series on ax1
ax1.set_xlabel('Wind speed (m/s)')
ax1.set_ylabel('Power (kW) - Turbine 1')
ax1.plot(X, Y1, label='Turbine 1')

# Create a second y-axis sharing the same x-axis
ax2 = ax1.twinx()
ax2.set_ylabel('Power (kW) - Turbine 2')
ax2.plot(X, Y2, label='Turbine 2')


# Set the colour of the tickmarks to show which axis they belong to
#ax1.tick_params(axis='y', colors='blue')
#ax2.tick_params(axis='y', colors='orange')

fig.suptitle('Plot with Separate Y Axes')
fig.legend()
plt.show()
```

**Subplots**

Our lines now overlap. What if the magnitudes are different but you instead want to plot them on different subplots?


```python
fig, (ax1, ax2) = plt.subplots(1, 2)  # Create a figure with 2 subplots side by side

# Plot first data series on the first subplot
ax1.plot(X, Y1)
ax1.set_xlabel('Wind speed (m/s)')
ax1.set_ylabel('Power (kW)')
ax1.set_title('Turbine 1')

# Plot second data series on the second subplot
ax2.plot(X, Y2)
ax2.set_xlabel('Wind speed (m/s)')
ax2.set_ylabel('Power (kW)')
ax2.set_title('Turbine 2')

fig.suptitle('Separate Subplots for Each Series')
fig.tight_layout() # Ensure there's no overlap between subplots
plt.show()
```

**Subplots with labelling**

It is good practice to have numerical or alphabetic labels for your subplots so that they can be easily referred to within the text. We can add the labels as part of the x axis labels

```python
fig, axs = plt.subplots(2, 2, figsize=(8, 8))  # Create a figure with 2 subplots side by side
ax1, ax2, ax3, ax4 = axs.flatten()

# Plot first data series on the first subplot
ax1.plot(X, Y1)
ax1.set_xlabel('Wind speed (m/s)\n\n' + r'$\mathbf{(a)~Turbine~1}$', labelpad=8)
ax1.set_ylabel('Power (kW)')

# Plot second data series on the second subplot
ax2.plot(X, Y2)
ax2.set_xlabel('Wind speed (m/s)\n\n' + r'$\mathbf{(b)~Turbine~2}$', labelpad=8)
ax2.set_ylabel('Power (kW)')

# Plot first data series on the third subplot
ax3.plot(X, Y1)
ax3.set_xlabel('Wind speed (m/s)\n\n' + r'$\mathbf{(c)~Turbine~3}$', labelpad=8)
ax3.set_ylabel('Power (kW)')

# Plot second data series on the fourth subplot
ax4.plot(X, Y2)
ax4.set_xlabel('Wind speed (m/s)\n\n' + r'$\mathbf{(d)~Turbine~4}$', labelpad=8)
ax4.set_ylabel('Power (kW)')

# Adjust the positioning of the subplots within the figure
fig.subplots_adjust(
    left=0.00,
    right=1.0,
    top=0.9,
    bottom=0.08,
    wspace=0.25,
    hspace=0.5
)

plt.show()
```

**Creating a function that will create a plot when called to**

You can also create functions that take your series values as inputs

```python
def plot_simple(X, Y, show_plot=True):
    fig, ax = plt.subplots()
    ax.plot(X, Y)
    ax.set_xlabel('Wind speed (m/s)')
    ax.set_ylabel('Power (kW)')
    ax.set_title('Power Curve')
    
    if show_plot:
        plt.show()
    else:
        plt.close(fig)
    return fig

def plot_subplots(X, Y1, Y2, show_plot=True):
    fig, (ax1, ax2) = plt.subplots(1, 2)  # Create a figure with 2 subplots side by side

    # Plot first data series on the first subplot
    ax1.plot(X, Y1)
    ax1.set_xlabel('Wind speed (m/s)')
    ax1.set_ylabel('Power (kW)')
    ax1.set_title('Turbine 1')

    # Plot second data series on the second subplot
    ax2.plot(X, Y2)
    ax2.set_xlabel('Wind speed (m/s)')
    ax2.set_ylabel('Power (kW)')
    ax2.set_title('Turbine 2')
    fig.suptitle('Separate Subplots for Each Series')
    fig.tight_layout()

    if show_plot:
        plt.show()
    else:
        plt.close(fig)
    return fig


simple_fig = plot_simple(X, Y, show_plot=True)
subplot_fig = plot_subplots(X, Y1, Y2, show_plot=True)
```
