**Import the matplotlib library:**

```python
import matplotlib.pyplot as plt # Import the matplotlib library
```

**Now let's create a figure:**

```python
X = [1, 2, 3] # Define X data
Y = [4, 5, 6] # Define Y data

fig, ax = plt.subplots()

ax.plot(X, Y)
plt.show()
```

**Now let's add some axis labels to our plot:**

```python
fig, ax = plt.subplots()

ax.set_xlabel('X Axis Label') # Set the name of the x-axis
ax.set_ylabel('Y Axis Label') # set the name of the y-axis

ax.plot(X, Y)
plt.show()
```

**Let's add a plot title:**

```python
fig, ax = plt.subplots()

ax.set_title('Plot Title')  # Set the title of the plot
ax.set_xlabel('X Axis Label') # Set the name of the x-axis
ax.set_ylabel('Y Axis Label') # set the name of the y-axis

ax.plot(X, Y)
plt.show()
```

**Finally, let's put it all together and save as an image:**

```python
fig, ax = plt.subplots()

X = [1, 2, 3] # Define X data
Y = [4, 5, 6] # Define Y data
ax.set_xlabel('X Axis Label') # Set the name of the x-axis
ax.set_ylabel('Y Axis Label') # set the name of the y-axis
ax.set_title('Plot Title')  # Set the title of the plot
ax.plot(X, Y)

plt.savefig('my_figure.jpeg', format='jpeg', dpi=200, bbox_inches='tight') # Save the figure as a JPEG file with specified resolution and bounding box with little whitespace
plt.show()
```

**What if we have data in a spreadsheet that we want to import?** 

We can import the 'pandas' library

```python
import pandas as pd # Import the pandas library (use 'pip install pandas openpyxl' to install the library)

df = pd.read_excel("example_data.xlsx") # Read data from an Excel file into a DataFrame

# If you want a specific sheet:
#df = pd.read_excel("example_data.xlsx", sheet_name="Sheet2")

print(df) # let's see what data we have

X = df["wind speed"] # Extract the "wind speed" column for X data
Y = df["power"]      # Extract the "power" column for Y data

fig, ax = plt.subplots()
ax.plot(X, Y)
ax.set_xlabel("Wind speed (m/s)")
ax.set_ylabel("Power (kW)")
ax.set_title("Power Curve")
plt.show()
```
