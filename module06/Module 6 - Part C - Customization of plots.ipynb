**Import the matplotlib library:**

```python
import matplotlib.pyplot as plt # Import the matplotlib library
from IPython.display import display # Import display function from IPython (so we can edit the figures as we go)
```

**Let's create a plot and customize the font sizes**

```python
fig = plt.figure(figsize=(7, 5), dpi=100) # figure size and resolution

ax = fig.add_subplot() # Add axes to the figure

X1 = [1, 2, 3] # Define X data
Y1 = [4, 5, 6] # Define Y data
Y2 = [7, 8, 9] # Define second Y data

ax.set_xlabel('X Axis Label', fontsize=14, labelpad=10)
ax.set_ylabel('Y Axis Label', fontsize=14, labelpad=10)

ax.tick_params(axis='x', labelsize=12)
ax.tick_params(axis='y', labelsize=12)

ax.set_title('Plot Title', fontsize=14, fontweight='bold')

# Add labels for the legend
ax.plot(X1, Y1, label='Series 1') 
ax.plot(X1, Y2, label='Series 2')

ax.legend(loc='upper left', frameon=True, fontsize=12, framealpha=1.0)  # Add and customize the legend
plt.show()
```

**Customizing font styles**

```python
# fontstyle='normal' | 'italic' | 'oblique'
# fontweight='light' | 'normal' | 'bold' | 'heavy'
# family='serif' | 'sans-serif' | 'monospace' | 'cursive' | 'fantasy'

X1 = [1, 2, 3] # Define X data
Y1 = [4, 5, 6] # Define Y data
Y2 = [7, 8, 9] # Define second Y data

fig = plt.figure(figsize=(7, 5), dpi=100) # figure size and resolution
ax = fig.add_subplot() # Add axes to the figure

# Customize fonts for title and labels individually
ax.set_title("Plot title", fontsize=16, fontstyle='italic', fontweight='bold')
ax.set_xlabel("X Axis label", fontsize=12, family='serif')
ax.set_ylabel("Y Axis label", fontsize=12, family='monospace')

ax.tick_params(axis='x', labelsize=12)
ax.tick_params(axis='y', labelsize=12)
ax.plot(X1, Y1, label='Series 1') 
ax.plot(X1, Y2, label='Series 2')

ax.legend(loc='upper left', frameon=True, fontsize=12, framealpha=1.0)  # Add and customize the legend
plt.show()

# Set global font properties
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.size'] = 14
plt.rcParams['font.style'] = 'italic'

fig = plt.figure(figsize=(7, 5), dpi=100) # figure size and resolution
ax = fig.add_subplot() # Add axes to the figure

ax.set_title("Plot title")
ax.set_xlabel("X Axis label")
ax.set_ylabel("Y Axis label")

ax.tick_params(axis='x', labelsize=12)
ax.tick_params(axis='y', labelsize=12)
ax.plot(X1, Y1, label='Series 1') 
ax.plot(X1, Y2, label='Series 2')
ax.legend(loc='upper left', frameon=True, fontsize=12, framealpha=1.0)  # Add and customize the legend
plt.show()
```

**Let's customize the colors of the lines.**

You can plot with many different colors by simply stating the name, e.g.: 'red', 'blue', 'orange', 'yellow', 'green', 'purple', 'black', 'brown', 'cyan', 'indigo', 'violet', 'gray' etc.

Or, precise colours can be specified by setting the (r,b,g) combination, with each value between 0-1.

```python
fig = plt.figure() # Create a figure object
ax = fig.add_subplot() # Add axes to the figure
ax.set_xlabel('X Axis Label') # Set the name of the x-axis
ax.set_ylabel('Y Axis Label') # set the name of the y-axis
ax.set_title('Different line colors')  # Set the title of the plot

# Customize the line styles for each series
ax.plot(X1, Y1, color='red', label='Series 1') # or ax.plot(X1, Y1, color=(1,0,0), label='Series 1')  or ax.plot(X1, Y1, color="0.5", label='Series 1')
ax.plot(X1, Y2, color='blue', label='Series 2') # or ax.plot(X1, Y1, color=(0,1,0), label='Series 1') or ax.plot(X1, Y1, color="0.7", label='Series 2')

ax.legend()
plt.show()
```

**Customizing the line styles of different series**

We can customize the line styles of our different series. This is set with the 'linestyle' option. The different linestyles that you can use and their syntaxes are:

`-`  Solid

`--` Dashes

`:`  Dots

`-.` Dash-dot

```python
fig = plt.figure() # Create a figure object
ax = fig.add_subplot() # Add axes to the figure
ax.set_xlabel('X Axis Label') # Set the name of the x-axis
ax.set_ylabel('Y Axis Label') # set the name of the y-axis
ax.set_title('Different line colors and styles')  # Set the title of the plot


X1 = [1, 2, 3] # Define X data

Y1 = [4, 5, 6] # Define Y data
Y2 = [7, 8, 9] # Define second Y data
Y3 = [10, 11, 12] # Define third Y data
Y4 = [13, 14, 15] # Define fourth Y data


# Customize the line styles for each series
ax.plot(X1, Y1, color='red', linestyle='-', label='Series 1')
ax.plot(X1, Y2, color='blue', linestyle='--', label='Series 2')
ax.plot(X1, Y3, color='green', linestyle=':', label='Series 3')
ax.plot(X1, Y4, color='orange', linestyle='-.', label='Series 4')

ax.legend()  # Update legend to reflect new labels
plt.show()
```

**Customize the line markers**

We can also add custom line markers to our data series. The different markers and their syntaxes are:

`.` Dot

`o`	Circle

`^`	Upward-pointing triangle

`v`	Downward-pointing triangle

`<`	Left-pointing triangle

`>`	Right-pointing triangle

`s`	Square

`p`	Pentagon

`*`	Star-shaped

`h`	Hexagon (pointy)

`H`	Hexagon (flat)

`+`	Plus

`x`	Cross

`D`	Diamond

`d`	Narrow diamond

`''`	No marker


```python
fig = plt.figure() # Create a figure object
ax = fig.add_subplot() # Add axes to the figure
ax.set_xlabel('X Axis Label') # Set the name of the x-axis
ax.set_ylabel('Y Axis Label') # set the name of the y-axis
ax.set_title('Lines with markers')  # Set the title of the plot

# Customize the line styles for each series
ax.plot(X1, Y1, color='red', marker='o', label='Series 1')
ax.plot(X1, Y2, color='blue', marker='*', label='Series 2')
ax.plot(X1, Y3, color='green', marker='s', label='Series 3')
ax.plot(X1, Y4, color='orange', marker='D', label='Series 4')

ax.legend()  # Update legend to reflect new labels
plt.show()
```

**Gridlines**

We can also add gridlines to our plots

```python
ax.grid(True)  # Add gridlines to the plot
ax.set_title('Gridlines')  # Set the title of the plot


fig.canvas.draw()  # Redraw the figure to show updates
display(fig)
```

We can also change the transparency and the linestyle of the gridlines

```python
ax.grid(True, linestyle='--', alpha=0.5)  # Dashed gridlines with some transparency
ax.set_title('Dashed gridlines')  # Set the title of the plot


fig.canvas.draw()
display(fig)
```

```python

ax.set_title('Axis limits and separations')  # Set the title of the plot
ax.set_aspect(aspect='auto')

# Set custom x and y axis limits
ax.set_xlim(1, 3)
ax.set_ylim(0, 16)

# Adjust the spacing between x points by changing the ticks
ax.set_xticks([1, 1.5, 2, 2.5, 3])
ax.set_yticks(range(0, 18, 2))

fig.canvas.draw()
display(fig)
```

```python
# Set tick marks to be inside, outside, or across the axes
ax.set_title('Tick marks inside')
ax.tick_params(axis='both', direction='in')      # Tick marks inside
fig.canvas.draw()
display(fig)

#To set tick marks outside:
ax.set_title('Tick marks outside')
ax.tick_params(axis='both', direction='out')
fig.canvas.draw()
display(fig)

# To set tick marks across (both in and out):
ax.set_title('Tick marks in and out')
ax.tick_params(axis='both', direction='inout')
fig.canvas.draw()
display(fig)

ax.tick_params(axis='both', direction='in')
ax.tick_params(axis='both', direction='out')
ax.tick_params(axis='both', direction='inout')
```
